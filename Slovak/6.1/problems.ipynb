## Úlohy na hodinu

## ⭐ Úloha 01

Napíš funkciu, ktorá berie ľubovoľný string a vráti počet číslic v tomto stringu.

## ⭐ Úloha 02

Napíš funkciu, ktorá vezme string a vráti ten istý string, ale písmená budú na striedačku malé a veľké (tAkTo).

## ⭐ Úloha 03

Vytvor funkciu, ktorá vezme dlhý string s viacerými riadkami a vráti list tohto dlhého reťazca rozdeleného podľa riadkov. Môžeš predpokladať, že riadky sú oddelené značkou zlomu riadku, teda `\n`. 

## ⭐ Úloha 04

Vytvor funkciu, ktorá vezme dlhý string s viacerými vetami a vráti zoznam viet, ktoré sa v tomto stringu nachádzajú.

## ⭐ Úloha 05

Začni s dlhým stringom (s viacerými vetami) a vytvor stĺpcový graf (bar chart), kde každý stĺpec bude predstavovať písmeno abecedy a bude počítať, koľko slov v danom stringu začína na toto písmeno. (Hint: Skús použiť špecifickú funkciu pre v regexe).
