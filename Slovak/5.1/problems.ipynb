# Úlohy na hodinu


## ⭐ Úloha 01
Máš dva listy s usporiadanými číslami. Spoj ich do jedného usporiadaného listu, pričom môžeš použiť len operácie pre rad (queue).

Príklad: `queue1 = [1, 3, 5]`, `queue2 = [2, 4, 6]` → `[1, 2, 3, 4, 5, 6]`


## ⭐ Úloha 02
Máš list čísel. Použi iba jeden stack a vráť ten istý list len naopak.

Príklad: `[1, 2, 3, 4, 5]` → `[5, 4, 3, 2, 1]`


## ⭐ Úloha 03
Máš usporiadaný list čísel. Použi jeden rad (queue) a jeden zásobník (stack) a vráť pole s nepárnymi číslami na pôvodných pozíciách a párnymi číslami otočenými.

Príklad: `[1, 2, 3, 4, 5]` → `[1, 4, 3, 2, 5]`


## ⭐ Úloha 04
Máš zoznam čísel `[3, 1, 4, 1, 5, 9]`. Použi iba zásobník (stack) na nájdenie najbližšieho menšieho prvku naľavo od každého čísla. 

Príklad: pre 4 je to 1; pre 5 je to 4.


## ⭐ Úloha 05
Máš binárny strom s hodnotami `[1, 2, 3, 4, 5]`. Napíš funkciu, ktorá vráti maximálnu hĺbku (počet úrovní) stromu.

## ⭐ Úloha 06
Máš min-heap `[1, 3, 5, 7, 9, 11]`. Vlož číslo 4 a vráť túto heap tak, aby bola zachovaná heap vlastnosť (najmenší prvok je navrchu).

## ⭐ Úloha 07
Vytvor triedu `Node` s atribútmi `value`, `left` a `right`. Vytvor binárny strom s koreňom 1, ľavým potomkom 2 a pravým potomkom 3. Vytlač všetky tri hodnoty.

## ⭐ Úloha 08
Vytvor binárny strom, ktorý vždy pridáva nové uzly do ľavej vetvy až do hĺbky 3, potom sa prepne na pravú. Začni s koreňovou hodnotou 1 a pridaj uzly 2, 3, 4, 5, 6, 7, 8.

## ⭐ Úloha 09
Implementuj Bubble Sort. Usporiadaj list `[64, 34, 25, 12, 22]` a spočítaj, koľko prehodení (swapov) bolo urobených. Vytlač usporiadaný list aj počet prehodení.

## ⭐ Úloha 10
Usporiadaj list stringov podľa abecedy pomocou Merge Sort.
