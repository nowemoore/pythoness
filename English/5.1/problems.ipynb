# In-Class Problems


## ⭐ Task 01
You're given two queues with sorted numbers. Merge them into one sorted list using only queue operations. 

Example: `queue1 = [1, 3, 5]`, `queue2 = [2, 4, 6]` → `[1, 2, 3, 4, 5, 6]`


## ⭐ Task 02
You're given a list of numbers. Using only one stack, reverse the order of all elements. 

Example: `[1, 2, 3, 4, 5]` → `[5, 4, 3, 2, 1]`


## ⭐ Task 03
You're given an ordered array of numbers. Using one queue and one stack, return an array with odd numbers in original positions and even numbers reversed. 

Example: `[1, 2, 3, 4, 5]` → `[1, 4, 3, 2, 5]`


## ⭐ Task 04
You're given a list of numbers `[3, 1, 4, 1, 5, 9]`. Using only a stack, find the nearest smaller element to the left of each number. 

Example: for 4, it's 1; for 5, it's 4.


## ⭐ Task 05
You're given a binary tree with values `[1, 2, 3, 4, 5]`. Write a function that returns the maximum depth (number of levels) of the tree.

## ⭐ Task 06
You have a min-heap `[1, 3, 5, 7, 9, 11]`. Insert the number 4 and return the heap maintaining the min-heap property (smallest element at top).

## ⭐ Task 07
Create a `Node` class with `value`, `left`, and `right` attributes. Build a binary tree with root 1, left child 2, right child 3. Print all three values.

## ⭐ Task 08
Create a binary tree that always adds new nodes to the left branch until depth 3, then switches to right. Starting with root value 1, add nodes 2, 3, 4, 5, 6, 7, 8.

## ⭐ Task 09
Implement bubble sort from scratch. Sort the list `[64, 34, 25, 12, 22]` and count how many swaps were made. Print both the sorted list and swap count.

## ⭐ Task 10
Sort a list of strings using Merge Sort.
