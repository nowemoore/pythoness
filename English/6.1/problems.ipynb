## In-Class Problems

## ⭐ Task 01

Write a function that takes in a string and returns the number of digits in this string.

## ⭐ Task 02

Write a function that takes in a string and returns the same string with every other letter lowercase and every other letter uppercase.

## ⭐ Task 03

Make a function that takes in a long string consisting of multiple lines and returns a list of this long string split by lines. You may assume that lines are split by line break marker `\n`. 

## ⭐ Task 04

Make a function that takes in a long string consisting of multiple lines and returns a list of sentences contained in this string.

## ⭐ Task 05

Begin with a long string (of multiple sentences) and make a bar chart where each bar represents a letter of alphabet and counts how many words in the given string begins with this letter. (Hint: consider using a specific regex function.)

## ⭐ Task 06

Find a dataset of Kaggle with numeric variables that you think may collectively predict something. Use unsupervised learning to discover natural groups in the data.
